```python
import librosa as lb
import matplotlib.pyplot as plt
import librosa.display
import soundfile as sf
from IPython.display import Audio
import numpy as np
from scipy.io.wavfile import write
import scipy.signal as sp
```

```python
# A1. Take a portion of your recorded signal which represents a vowel sound. Perform FFT on the signal snippet and observe the amplitude spectrum.
audio='vowel.wav'
signal1,samprate=lb.load(audio,sr=None)

print("Vowel Signal:")
Audio(data=signal1, rate=samprate)
```

```python
FT=np.fft.fft(signal1)
```

```python
plt.figure(figsize=(10,5))
plt.plot(signal1,color='indigo')
plt.title('Vowel Speech')
plt.show()

plt.figure(figsize=(10,5))
plt.plot(np.real(FT),color='violet')
plt.title("Amplitude Spectrum")
plt.show()
```

```python
# A3. Repeat A2 for few slices of silence & non-voiced portions of the recorded speech signal
audio='AI-in-Speech-processing.wav'
signal,samprate=lb.load(audio,sr=None)

print('Silence Signal')
Audio(signal[0:2000],rate = samprate)
```

```python
silence = signal[0:2000]
silence_fft = np.fft.fft(silence)

plt.figure(figsize=(20,5))
plt.plot(np.real(silence_fft),color='red')
plt.title('FFT of silence part')
```

```python
# non-voiced portions of the recorded speech signal
print('Non voiced Signal')
Audio(signal[4000:15000],rate = samprate)
```

```python
non_voice = signal[4000:15000]
non_voiced_fft = np.fft.fft(non_voice)

plt.figure(figsize=(20,5))
plt.plot(np.real(non_voiced_fft),color='indigo')
plt.title('FFT of non_voiced Part')
```

```python
# A4. Generate the spectrogram of the signal and observe the change points of the signals with associated speech segments
def generate_spectrogram(signal, sample_rate):
    plt.specgram(signal, Fs=sample_rate)
    plt.title("Spectrogram of Speech Signal")
    plt.xlabel("Time (s)")
    plt.ylabel("Frequency (Hz)")
    plt.colorbar()
```

```python
# Generating spectrogram
plt.figure(figsize=(12, 6))
generate_spectrogram(signal,samprate)
plt.show()
```
