```python
import librosa as lb
import matplotlib.pyplot as plt
import librosa.display
import soundfile as sf
from IPython.display import Audio
```

```python
audio='AI-in-Speech-processing.wav'
signal,samprate=lb.load(audio)
print("Sample Rate:",samprate)
```

```python
plt.figure(figsize=(10, 5))
lb.display.waveshow(signal,sr=samprate,color='green')
plt.title('Speech Signal')
plt.show()
```

```python
output_file1='resample1.wav'
target_srate=18000

resampled_signal = lb.resample(signal, orig_sr=samprate, target_sr=target_srate)
print(lb.get_duration(y=resampled_signal, sr=target_srate))
plt.figure(figsize=(10, 5))
lb.display.waveshow(resampled_signal, sr=target_srate, color='green')
plt.title('Resampled Speech Signal')
plt.show()
sf.write(output_file1,signal,target_srate,format='wav')
```

```python
audio_data, sr = lb.load(output_file1, sr=target_srate)
Audio(data=audio_data, rate=sr)
```

```python
output_file2='resample2.wav'
target_srate=30000

resampled_signal = lb.resample(signal,orig_sr=samprate,target_sr=target_srate)
print(lb.get_duration(y=resampled_signal, sr=target_srate))
plt.figure(figsize=(10,5))
lb.display.waveshow(resampled_signal,sr=target_srate, color='green')
plt.title('Resampled Speech Signal')
plt.show()
sf.write(output_file2,signal,target_srate,format='wav')
```

```python
audio_data, sr = lb.load(output_file2, sr=target_srate)
Audio(data=audio_data, rate=sr)
```

```python
reverse=signal[::-1]
output_file3='reversed.wav'
target_srate=samprate
plt.figure(figsize=(10,5))
lb.display.waveshow(reverse,sr=target_srate,color='green')
plt.title('Reversed Speech Signal')
sf.write(output_file3,reverse,target_srate,format='wav')
```

```python
audio_data, sr = lb.load(output_file3, sr=target_srate)
Audio(data=audio_data, rate=sr)
```
