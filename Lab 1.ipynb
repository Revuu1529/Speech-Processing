```python
import librosa as lb
import matplotlib.pyplot as plt
import librosa.display
import soundfile as sf
from IPython.display import Audio
```

```python
aud='AI-in-Speech-processing.wav'
signal,samprate=lb.load(aud)
print("Sample Rate:",samprate)
```

```python
plt.figure(figsize=(10, 4))
lb.display.waveshow(signal, sr=samprate)
plt.title('Speech Signal')
plt.show()
```

```python
output_file_path = 'resample1.wav'
target_sr = 18000

resampled_signal = lb.resample(signal,orig_sr=samprate, target_sr=target_sr)
print(lb.get_duration(y=resampled_signal, sr=target_sr))
plt.figure(figsize=(10, 4))
lb.display.waveshow(resampled_signal, sr=target_sr)
plt.title('Resampled Speech Signal')
sf.write(output_file_path, signal, target_sr, format='flac', subtype='PCM_24')
```

```python
output_file_path = 'resample2.wav'
target_sr = 25000

resampled_signal = lb.resample(signal,orig_sr=samprate, target_sr=target_sr)
print(lb.get_duration(y=resampled_signal, sr=target_sr))
plt.figure(figsize=(10, 4))
lb.display.waveshow(resampled_signal, sr=target_sr)
plt.title('Resampled Speech Signal')
sf.write(output_file_path, signal, target_sr, format='flac', subtype='PCM_24')
```

```python
reversed_sign=signal[::-1]
output_file_path = 'reversed.wav'
target_sr = samprate
plt.figure(figsize=(10, 4))
lb.display.waveshow(reversed_sign, sr=target_sr)
plt.title('Reversed Speech Signal')
sf.write(output_file_path, reversed_sign, target_sr, format='flac', subtype='PCM_24')
```
